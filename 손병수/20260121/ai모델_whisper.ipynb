```python
from google.colab import drive
import os
import shutil

drive.mount('/content/gdrive')

# --- 설정 구간 ---
# filepath = "본인의_음성파일_경로.mp3"  # 사용자가 이미 정의했다고 가정
# ----------------
filepath = '/content/gdrive/MyDrive/통화 녹음 01056188855_260115_160300.m4a'

# 3. Whisper 실행 (medium -> large-v3로 변경)
# --output_dir . 옵션을 추가하여 현재 폴더에 결과가 생성되도록 명시합니다.
!whisper "{filepath}" --model medium --language Korean --output_dir .

# 4. 결과 파일 처리 및 구글 드라이브 저장
filename = os.path.splitext(os.path.basename(filepath))[0]
colab_txt_path = f"{filename}.txt"
gdrive_folder = "/content/gdrive/MyDrive/test"
gdrive_txt_path = os.path.join(gdrive_folder, f"{filename}.txt")

# 폴더가 없으면 생성 (파일이 깨지거나 저장 오류가 나는 주요 원인 해결)
if not os.path.exists(gdrive_folder):
    os.makedirs(gdrive_folder)
    print(f"폴더 생성됨: {gdrive_folder}")

# Whisper가 파일을 정상적으로 생성했는지 확인 후 복사
if os.path.exists(colab_txt_path):
    # 'shutil.copy'를 사용하면 파일 쓰기 과정에서의 인코딩 오류를 방지하고 원본을 안전하게 복사합니다.
    shutil.copy(colab_txt_path, gdrive_txt_path)

    # 확인을 위해 내용을 출력 (선택 사항)
    with open(gdrive_txt_path, "r", encoding="utf-8") as f:
        print("\n--- 변환 결과 미리보기 ---")
        print(f.read()[:500] + "...") # 앞부분 500자만 출력

    print(f"\n 저장 완료: {gdrive_txt_path}")
else:
    print(f"에러: {colab_txt_path} 파일이 생성되지 않았습니다. Whisper 실행 로그를 확인하세요.")
```
